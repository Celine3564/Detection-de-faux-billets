# Partie 2 : Traitement des valeurs manquantes avec regression linéaire

## Régression linéaire simple de la variable margin_low et Lenght

### Verification des hypothèses de validité de la régression linéaire simple


```python
reg_simp = smf.ols('length ~ margin_low', data=billets_df).fit()
```

```python
print(reg_simp.summary())
```

```python
# Regression linéraire simple entre margin low et lenght

r_squared = reg_simp.rsquared
print('Ici la r_squared est égale à {}'.format(r_squared))


pvalue = reg_simp.f_pvalue
print('Ici la p-value est égale à {}'.format(pvalue))

# P-value : choix d'un seuil de probabilité standard, de 5%
seuil = 0.05

#Conclusion
if pvalue < seuil:
    print("On rejette HO : il y a une corrélation entre lenght et margin_low.")
elif pvalue > seuil : 
    print("On ne peut pas rejeter HO: il n'y a pas de corrélation entre lenght et margin_low")
    
    
```

Le $R^{2}$, il est de l'ordre de 44.4%. Ce n'est pas très élevé, mais ceci est logique au vu de la dispersion du nuage de points originel.



### Visualisez la droite de régression

Alors, voyons à quoi ressemble notre droite.

```python
ax = sns.lmplot(x="length", y="margin_low", data=billets_df, ci=None, line_kws={'color':'black'})
ax.set(xlabel='length', ylabel='Mmargin_lowxO3')
plt.show()
```

## regression linéraire multiples

On veut appliquer la régression linéaire pour modéliser la dimension margin_low des valeures manquantes en fonction de toutes les autres variables.
La régression linéaire suppose une relation linéaire entre les variables indépendantes et la variable dépendante.


```python
reg_multi = smf.ols('margin_low~diagonal+height_left+height_right+margin_up+length', data=billets_df).fit()
print(reg_multi.summary())
```

## Analyse des valeur abbertantes et influentes

### Calculez les leviers

On peut calculer les leviers comme ceci, en sachant que le seuil des leviers est de $2∗\frac{p}{n}$.

```python
alpha = 0.05
```

```python
n = billets_df.shape[0]
p = 6
```

```python
analyses = billets_df.copy()
analyses
```

```python
analyses['levier'] = reg_multi.get_influence().hat_matrix_diag

seuil_levier = 2*p/n
seuil_levier
```

```python
plt.figure(figsize=(15,6))
plt.bar(analyses.index, analyses['levier'])
plt.xticks(np.arange(0, 1500, step=100))
plt.xlabel('Observation')
plt.ylabel('Leviers')
plt.plot([0, 1500], [seuil_levier, seuil_levier], color='r')
plt.show()
```

Pour sélectionner les points pour lesquels le levier est supérieur au seuil, on exécute la ligne suivante :

```python
analyses.loc[analyses['levier'] > seuil_levier, : ]
```

il y a 75 valeurs supérieures au seuil à supprimer

```python
# Get names of indexes for which column Stock has value No
indexNames = analyses[ analyses['levier'] > seuil_levier ].index
# Delete these row indexes from dataFrame
analyses.drop(indexNames , inplace=True)
analyses
```

```python
reg_multi = smf.ols('margin_low~diagonal+height_left+height_right+margin_up+length', data=analyses).fit()
print(reg_multi.summary())
```

Le $R^{2}$, il est de l'ordre de 49.1%. 
augmentation de 2% suis au retrait des leviers

### Calculez les résidus studentisés

Si l'on souhaite maintenant calculer les résidus studentisés, nous écrivons ceci, sachant que le seuil pour les résidus studentisés est une loi de Student à n-p-1 degrés de liberté :

```python
n = analyses.shape[0]
p = 6
```

```python
analyses['rstudent'] = reg_multi.get_influence().resid_studentized_internal
seuil_rstudent = t.ppf(1-alpha/2,n-p-1)
```

```python
analyses
```

```python
plt.figure(figsize=(10,6))
plt.bar(analyses.index, analyses['rstudent'])
plt.xticks(np.arange(0, 1500, step=100))
plt.xlabel('Observation')
plt.ylabel('Résidus studentisés')
plt.plot([0, 1500], [seuil_rstudent, seuil_rstudent], color='r')
plt.plot([0, 1500], [-seuil_rstudent, -seuil_rstudent], color='r')
plt.show()
```

```python
analyses.loc[analyses['rstudent'] > seuil_rstudent, :].shape
```

```python
# Get names of indexes for which column Stock has value No
indexNames = analyses[ analyses['rstudent'] > seuil_rstudent ].index
# Delete these row indexes from dataFrame
analyses.drop(indexNames , inplace=True)
analyses
```

```python
reg_multi = smf.ols('margin_low~diagonal+height_left+height_right+margin_up+length', data=analyses).fit()
print(reg_multi.summary())
```

Le $R^{2}$, il est de l'ordre de 52,3%. 
augmentation de 3% suis au retrait des residus

### Déterminez la distance de Cook

Pour trouver la distance de Cook, nous exécutons ceci :

```python
#recalcul du nombre de lignes
n = analyses.shape[0]
p = 6
```

```python
influence = reg_multi.get_influence().summary_frame()
```

```python
analyses['dcooks'] = influence['cooks_d']
seuil_dcook = 4/(n-p)
```

On peut détecter les observations influentes comme ceci :

```python
plt.figure(figsize=(10,6))
plt.bar(analyses.index, analyses['dcooks'])
plt.xticks(np.arange(0, 1500, step=100))
plt.xlabel('Observation')
plt.ylabel('Leviers')
plt.plot([0, 1500], [seuil_dcook, seuil_dcook], color='r')
plt.show()
```

```python
analyses.loc[analyses['dcooks'] > seuil_dcook, :].shape
```

On ne retire des points qu'après avoir vérifié qu'ils sont effectivement atypiques, voire aberrants, au vu du modèle estimé.

```python
# Get names of indexes for which column Stock has value No
indexNames = analyses[ analyses['dcooks'] > seuil_dcook ].index
# Delete these row indexes from dataFrame
analyses.drop(indexNames , inplace=True)
analyses
```

```python
reg_multi = smf.ols('margin_low~diagonal+height_left+height_right+margin_up+length', data=analyses).fit()
print(reg_multi.summary())
```

Le $R^{2}$, il est de l'ordre de 56,1%. 
augmentation de 4% suis au retrait des distances cook

### Vérifier la colinéarité des variables

Une autre chose à vérifier est l'éventuelle colinéarité approchée des variables :

```python
variables = reg_multi.model.exog
[variance_inflation_factor(variables, i) for i in np.arange(1,variables.shape[1])]
```

Ici, tous les coefficients sont inférieurs à 10, il n'y a donc pas de problème de colinéarité.

### Testez l’homoscédasticité

On peut également tester l’homoscédasticité (c'est-à-dire la constance de la variance) des résidus :

Var(εi)=σ2  (leur variance est constante et égale à σ2 ) <br>
Les variances des résidus doivent être constantes. Les tests de Breusch-Pagan et de White permettent de tester l'hypothèse d'homoscédasticité du terme d'erreur d'un modèle de régression linéaire.<br>
(H0): Homoscédasticité: (les résidus sont dispersés aléatoirement, la variance d'un individu à un autre est constante.<br>
(Ha): hétéroscédasticité<br> 

```python
_, pval, __, f_pval = statsmodels.stats.diagnostic.het_breuschpagan(reg_multi.resid, variables)
print('p value test Breusch Pagan:', pval)
```

```python
# P-value : choix d'un seuil de probabilité standard, de 5%
seuil = 0.05

#Conclusion
if pval < seuil:
    print("On rejette HO : les résidus sont dispersés aléatoirement, la variance d'un individu à un autre est constant")
elif pval > seuil : 
    print("hétéroscédasticité")
```

## Test simple à 2 variables

Comme nous avons pu le voir dans l'étude exploratoire des données l'etude des données Lenght et Margin low permettent déjà de faire une 1ere etude de la probailité qu'un billet soit vrai ou faux

```python
fig, ax = plt.subplots(1,2, figsize=(15,8))
sns.boxplot(x='is_genuine', y='margin_low',data=df, ax=ax[0])
sns.boxplot(x="is_genuine", y="length", data=df,ax=ax[1])

plt.show()
```

```python
# choix visuel des valeurs de paramètres
min_margin_low = 4.5
max_lenght = 112.5
```

```python
#fonction afo, de predir le résultat
def f(row):
    if row['margin_low'] > min_margin_low and row['length'] < max_lenght:
        val = 0
    else:
        val = 1
    return val

#Création d'une nouvelle colonne
df['test simple'] = df.apply(f, axis=1)

#visualisation du data frame 
df
```

```python
#Représentation des effectifs par type de billet vrai/faux
effectif = df['test simple'].value_counts()
effectif
```

```python
#selection des faux billets : verfication du nombre de lignes
faux_billet = df.loc[((df['margin_low'] > min_margin_low) & (df['length'] < max_lenght))]
faux_billet
```

### Préparation des données pour le test 

```python
# Récupération des valeurs de "is genuine"
y_data_origine = df['is_genuine'].copy()
y_data_origine
```

```python
# Récupération des valeurs de "is genuine"
y_basic =df['test simple'].copy()
y_basic
```

```python
# Création du dataframe
df_test = pd.concat([y_data_origine, y_basic], axis=1)
```

```python
# Création d'un colonne comparant les cluster kmeans avec is_genuine
df_test['result'] = df_test['test simple'] == df_test['is_genuine']
```

### Matrice confusion - modèle 'simple'

##  K-means (clustering non-supervisé)

```python
data = df.drop('is_genuine', axis=True).copy()
data = data.drop('test simple', axis=True).copy()
#data = df.drop('is_genuine', axis=True).copy()
X = data.values
X
```

```python
km = KMeans(n_clusters=2) # 2 clusters (Vrai billet VS faux billet)
km.fit(X)
clusters = km.labels_

data['cluster_km'] = clusters
```

```python
# Séparation de 'is_genuine'
y = data.loc[:,['cluster_km']].values
# Standardisation des features
x = StandardScaler().fit_transform(X)

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['F1', 'F2'])

df_kmeans = pd.concat([principalDf, data[['cluster_km']]], axis = 1)

df_kmeans
```

The recall is the ratio tp / (tp + fn) where tp is the number of true positives and fn the number of false negatives. The recall is intuitively the ability of the classifier to find all the positive samples.
